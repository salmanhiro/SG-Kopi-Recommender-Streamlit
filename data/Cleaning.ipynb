```python
from tqdm import tqdm

days = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]

for row in tqdm(range(len(df_sg_coffee))):
    json_ex = json.loads(df_sg_coffee["opening_hours"].iloc[row])
    print(json_ex)
    for day in days:
        hours = json_ex[day].split("-")
        print(hours)
        if hours != ['Closed']:
            try:
                df_sg_coffee[day + "_open"].iloc[row] = datetime.datetime.strptime(hours[0], '%H:%M')
                df_sg_coffee[day + "_close"].iloc[row] = datetime.datetime.strptime(hours[1], '%H:%M')
            except ValueError:
                df_sg_coffee[day + "_open"].iloc[row] = datetime.datetime.strptime(hours[0].split(' ')[0], '%H:%M')
                df_sg_coffee[day + "_close"].iloc[row] = datetime.datetime.strptime(hours[1].split(' ')[0], '%H:%M')
        else:
            df_sg_coffee[day + "_open"].iloc[row] = None
            df_sg_coffee[day + "_close"].iloc[row] = None
```

```python
df_pa = pd.read_csv("sg_PA.csv")
```

```python
df_pa.columns = ['Name', 'Malay', 'Chinese', 'Pinyin', 'Tamil', 'Region','Area', 'Population', 'Density']
```

```python
df_pa["lat"] = 0
df_pa["lon"] = 0
```

```python
# try to get location

import requests
import urllib.parse

for i in range (len(df_pa)):
    index = df_pa["Name"].iloc[i]
    address = f"{index}, Singapore"
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'

    response = requests.get(url).json()
    print(address)
    print(response[0]["lat"])
    print(response[0]["lon"])
    df_pa["lat"].iloc[i] = response[0]["lat"]
    df_pa["lon"].iloc[i] = response[0]["lon"]
```
